```python
import pandas as pd

# Load training and test datasets
df_train = pd.read_csv("car_insurance_premium_dataset.csv")
df_test = pd.read_csv("car_insurance_premium_dataset_TEST.csv")
```

```python
# Shape of the datasets
print("Train shape:", df_train.shape)
print("Test shape:", df_test.shape)

# Info summary
print("\nTrain Info:")
df_train.info()

print("\nTest Info:")
df_test.info()

# Check for missing values
print("\nMissing values in Train:\n", df_train.isnull().sum())
print("\nMissing values in Test:\n", df_test.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Simulate time (e.g., Day 1 to Day 1000)
df_train['Day'] = range(1, len(df_train) + 1)

plt.figure(figsize=(10, 5))
plt.plot(df_train['Day'], df_train['Insurance Premium ($)'], color='blue')
plt.title("Insurance Premium Trend Over Time")
plt.xlabel("Day")
plt.ylabel("Insurance Premium ($)")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

avg_premium_by_accidents = df_train.groupby("Previous Accidents")["Insurance Premium ($)"].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=avg_premium_by_accidents, x="Previous Accidents", y="Insurance Premium ($)", palette="mako")
plt.title("Average Insurance Premium by Number of Previous Accidents")
plt.xlabel("Previous Accidents")
plt.ylabel("Average Premium ($)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_train["Insurance Premium ($)"], bins=30, kde=True, color="orange")
plt.title("Distribution of Insurance Premiums")
plt.xlabel("Insurance Premium ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_train, x="Driver Age", y="Insurance Premium ($)", hue="Previous Accidents", palette="viridis")
plt.title("Insurance Premium vs Driver Age")
plt.xlabel("Driver Age")
plt.ylabel("Insurance Premium ($)")
plt.legend(title="Previous Accidents")
plt.show()
```
